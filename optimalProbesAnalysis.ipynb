# Analysis
Below are certain plots to understand the output generated from the program.

```python
%matplotlib inline
import matplotlib.pylab as plt
```

```python
populationSize=24
N_ITERATIONS=10

scores=[]

for i in range(1,N_ITERATIONS+1):
        f= open('iter_'+str(i)+'_output_sets.txt','rb')
        for line in f:
            columns=line.strip('\n').split('\t')
            scores.append(columns[2])
        f.close()
        
t=[]
for i in range(0,N_ITERATIONS):
    l=i*populationSize
    u=l+populationSize
    t.append(l)
    x=range(l,u)
    plt.plot(x,scores[l:u],'o')
```

Note that the scores are increasing as the number of rounds are progressing.
